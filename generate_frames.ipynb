```python
import os
import cv2

# Path to the main folder containing videos
main_folder = path_videos

# Loop through each class folder
for class_name in os.listdir(main_folder):
    class_folder = os.path.join(main_folder, class_name)
    
    # Loop through each video in the class folder
    for video_id in os.listdir(class_folder):
        video_path = os.path.join(class_folder, video_id)
        
        # Open the video using OpenCV
        video = cv2.VideoCapture(video_path)
        
        # Create a folder to store frames for each video
        frames_folder = os.path.join(class_folder, f"{video_id}_frames")
        os.makedirs(frames_folder, exist_ok=True)
        
        # Extract frames from the video
        frame_count = 0
        while True:
            success, frame = video.read()
            if not success:
                break
            
            # Save the frame as an image file
            frame_path = os.path.join(frames_folder, f"frame_{frame_count}.jpg")
            cv2.imwrite(frame_path, frame)
            
            frame_count += 1
        
        # Release the video object
        video.release()
        print(f"Frames extracted from video: {video_path}")
        # break

print("Frame extraction complete.")
```
